```python
%%file task.py

import csv
from mrjob.job import MRJob

class MRTask(MRJob):
    def mapper_init(self):
        self.replicas_count = dict() 

    def mapper_final(self):
        for key, value in self.replicas_count.items():
            yield (key, value)
    
    def mapper(self, _, line):
        for result in csv.reader([line], delimiter=' '):
            if len(result) != 3:
                continue
            actor = result[1]
            self.replicas_count[actor] = self.replicas_count.get(actor, 0) + 1

    def reducer(self, key, values):
        yield (key, sum(values))

if __name__ == '__main__':
    MRTask.run()
```

```python
import csv
import matplotlib.pyplot as plt

filepath4 = '/data/SW_EpisodeIVResult/part-00000'
filepath5 = '/data/SW_EpisodeVResult/part-00000'
filepath6 = '/data/SW_EpisodeVIResult/part-00000'
filepathf = '/data/SWResult/part-00000'

def calculate(filepath, title):
    with hdfs.open_input_stream(filepath) as file:
        replicas_count = list()
        
        full = file.readall().decode('utf-8').strip().split('\n')
        for raw in full:
            actor, number = raw.split('\t')[0], int(raw.split('\t')[1])
            replicas_count.append((number, actor))
    
        replicas_count = sorted(replicas_count, reverse=True)
        replicas_count = replicas_count[:20]
        
        plt.bar([it[1] for it in replicas_count], [it[0] for it in replicas_count])
        plt.title(title, fontsize=14)
        plt.xlabel('Actor', fontsize=14)
        plt.xticks(rotation=90)  # Rotate x-axis labels vertically
        plt.ylabel('Number', fontsize=14)
        plt.grid(True)
        plt.show()

calculate(filepath4, '4')
calculate(filepath5, '5')
calculate(filepath6, '6')
calculate(filepathf, 'F')
```
